```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = sns.load_dataset('penguins')
sns.pairplot(df)
```

```python
tip = sns.load_dataset('tips')
tip.head()
```

```python
sns.relplot(x='total_bill' , y='tip' , col='time' , hue='smoker' , size="size" , data=tip)
```

```python
sns.lmplot(x='total_bill' , y='tip' , data=tip)
```

```python
sns.lmplot(x='size' , y='tip' , data=tip)
```

```python
sns.lmplot(x='size' , y='tip' , data=tip , x_estimator = np.mean)
```

```python
tip['big_tip'] = (tip.tip / tip.total_bill) > 0.15
sns.lmplot(x='total_bill' , y='big_tip' , y_jitter=0.03 , logistic=True , data=tip)
```

```python
sns.jointplot(x='total_bill' , y='tip' , kind='reg' , data=tip)
```
