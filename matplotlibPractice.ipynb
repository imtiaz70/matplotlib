```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

x = np.linspace(-3,3,30)
y = x**2
# plt.plot(x,y,'g3')
fig=plt.figure()
ax = fig.add_axes([100,1000,1.5,1])
ax.legend()
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x),'r-')
plt.plot(x,-np.sin(x),'g--')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# plt.subplot(nrows,ncols)
fig,a = plt.subplots(2,2)
x = np.arange(1,5)
print(x)

a[0][0].plot(x,x*x)
a[0][0].set_title("Sqare")

a[0][1].plot(x, np.sqrt(x))
a[0][1].set_title("Sqare Root")

a[1][0].plot(x,np.exp(x))
a[1][0].set_title("Exp")

a[1][1].plot(x,np.log10(x))
a[1][1].set_title("Log 10")

# Un Comment it
# plt.tight_layout() 
```

```python
a1 = plt.subplot2grid((3,3),(0,0),colspan=2)
a2 = plt.subplot2grid((3,3),(0,2),rowspan=2)
a3 = plt.subplot2grid((3,3),(1,0),rowspan=2 , colspan=2)

import numpy as np
x = np.arange(1,10)

a2.plot(x,x*x)
a2.set_title("Sqare")

a1.plot(x,np.exp(x))
a1.set_title("exp")

a3.plot(x,np.log(x))
a3.set_title("log")

plt.tight_layout()
plt.show()
```
