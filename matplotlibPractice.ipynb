```python
import matplotlib.pyplot as plt
import numpy as np

girls_grades = [80,89,90,70,87]
boys_grades = [30,50,60,70,99]
grades_range = [20,30,40,33,100]

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.scatter(grades_range,girls_grades,color='g')
ax.scatter(grades_range,boys_grades , color='r')
ax.set_xlabel('Grades range')
ax.set_ylabel('Grade Score')
ax.set_title('Scatter Plot')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

xlist = np.linspace(-3,3,100)
ylist = np.linspace(-3,3,100)

X , Y = np.meshgrid(xlist,ylist)
Z = np.sqrt(X**2 + Y**2)

fig , ax = plt.subplots(1,1)
cp = ax.contourf(X,Y,Z)

# Add color bar to plot
fig.colorbar(cp)
ax.set_title('Filled counter plot')
ax.set_xlabel('x(cm)')
ax.set_ylabel('y(cm)')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x,y = np.meshgrid(np.arange(-2,2,.2) , np.arange(-2,2,.25))
z = x*np.exp(-x**2 -y**2)
v,u = np.gradient(z , .2 , .2)
fig , ax = plt.subplots()
q = ax.quiver(x,y,u,v)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

collegesData = [[30,25,50,20],[40,30,51,17],[35,22,41,19]]
X = np.arange(4)
print(X)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.bar(X + 0.00 , collegesData[0] , color='b' , width=0.25 )
ax.bar(X + 0.25 , collegesData[1] , color='g' , width=0.25 ) 
ax.bar(X + 0.50 , collegesData[2] , color='r' , width=0.25 )
ax.set_xticks(X,['Punjab College','Superior college','Govt College','chinaab College'])
ax.legend(['SE','CS','IT'])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
N = 5
womenMean = (20,35,30,35,27)
menMeans = (25,32,34,20,25)
ind = np.arange(N)
width = 0.35
fig = plt.figure()
ax = fig.add_axes([1,1])
ax.bar(ind , menMeans,width,color='r')
ax.bar(ind,womenMean,width,bottom=menMeans,color='b')
ax.set_title('scores by group and gender')
ax.set_ylabel('Scores')
ax.set_xticks(ind,('G1','G2','G3','G4','G5'))
ax.set_yticks(np.arange(0,81,10))
plt.show()
```
