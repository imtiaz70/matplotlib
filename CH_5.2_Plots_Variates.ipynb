```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("googleplaystore.csv")
# df.head()
plt.style.use('dark_background')  # dark_background
sns.distplot(df.Rating , bins=20 , kde = False , color='r')
plt.show()
```

```python
df['Content Rating'].value_counts()
```

```python
df = df[~df['Content Rating'].isin([ 'Adults only 18+' , 'Unrated' ])] 
df.reset_index(inplace=True , drop=True)
df['Content Rating'].value_counts()
```

```python
plt.figure(figsize=[9,7])
df['Content Rating'].value_counts().plot.pie()
```

```python
plt.figure(figsize=[9,7])
df['Content Rating'].value_counts().plot.bar()
```
