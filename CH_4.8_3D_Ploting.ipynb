# Three dimensional plotting 

```python
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

z = np.linspace(0,1,100)
x = z * np.sin(20*z)
y = z * np.cos(20*z)
c = x + y
ax.plot3D(x,y,z)
ax.set_title('3D line Plot')
plt.show()
```

# 3D Contour Plot

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x,y):
    return np.sin(np.sqrt(x**2 + y**2))

x = np.linspace(-6,6,30)
y = np.linspace(-6,6,30)
x , y = np.meshgrid(x,y)
z = f(x,y)

fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.contour3D(x,y,z,50 , camp='binary')

# ax.view_init(60,35)                       Uncomment This

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Countour')

plt.show()
```

```python
from turtle import color
import numpy as np
import matplotlib.pyplot as plt

def f(x,y):
    return np.sin(np.sqrt(x**2 + y**2))

x = np.linspace(-6,6,30)
y = np.linspace(-6,6,30)
x , y = np.meshgrid(x,y)
z = f(x,y)

fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.plot_wireframe(x,y,z , color='green')

# ax.view_init(60,35)                     #  Uncomment This

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Wireframe')

plt.show()
```

# 3D Surface Plot

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.outer(np.linspace(-2,2,30),np.ones(30))
y = x.copy().T                                  # Transpose
z = np.cos(x**2 + y**2)

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x,y,z)

ax.set_title('Surface plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')


plt.show()
```
